### GroupBy

#### Univariate

```python
# which cooling type is more efficient
Final_df.groupby('coolingtype').count()
```

```python
# which product has more reliable (i.e,which has more compressor type that is efficient)
Final_df.groupby('Compressor_type').count()
```

### Observation:
   *Digital Inverter is the most frequent use of all compressor type*

#### Bivariate

```python
#Which product is availiable under their budget prices?
Final_df.groupby('Brand')[['Price']].max()
Final_df.groupby('Brand')[['Price']].min()
Final_df.groupby('Brand')[['Price']].mean()
```

### Observation:
  *Maximum, minimum and average price of all the brands*

```python
# which brand has highest rating
Final_df.groupby('Brand')['Rating'].max()
```

### Observation:
   *Maximum Rating of all brands. **Godrej,Haier,Samsung,Whirlpool** has highest rating.*

### Multi-variate

```python
# average rating of all brands Doortype
Final_df.groupby(['Brand','Door'])['Rating'].mean()
```

### Observation:
   *Average rating of all brands with Doortype. **Hisense(side by side door)** has highest average rating.*

```python
Final_df.groupby(['Brand','coolingtype'])['Litres_capacity(in litres)'].mean()
```

### Observation:
   Average Litres capacity of all brands including cooling type.The highest litres capacity is **Hisense(Frost free)** is 609L

### Observation:

### Pivot tables

#### Univariate

```python
pd.pivot_table(data=Final_df,index=['Brand'],aggfunc=np.mean)
```

### Observation:

#### Bivariate

```python
#get the highest price of the brand
pd.pivot_table(data=Final_df,index=['Brand'],values='Price',aggfunc=np.max)
```

### Observation:

```python
#average compressor warrenty of particular brand
pd.pivot_table(data=Final_df,index=['Brand'],values=['Compressor_Warranty(in years)'],aggfunc=np.mean)
```

### Observations:
**Bosch,CANDY,TOSHIBA,Sansui,Motorola,Lifelong,Hisense,Hitachi**  offering *10 years of compressor warranty*

#### Multivariate

```python
# the average price of all brands with door type using pivot 
pd.pivot_table(data=Final_df,index=['Brand','Door','coolingtype'],values='Price',aggfunc=np.mean)
```
